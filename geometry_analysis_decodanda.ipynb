```python
# Imports
import pickle
from decodanda import *
np.random.seed(0)
```

```python
# Analyze visual data
session_VIS = pickle.load(open('./datasets/VISp.pck', 'rb'))

dec = Decodanda(session_VIS, {'stimulus': [-1, 1], 'action': [-1, 1]}, verbose=False)
visualize_PCA(dec, mean=True, names=['A>', 'A<', 'B>', 'B<'], ndata=200)
plt.suptitle('Visual area')
plt.savefig('./img/visual_pca.png')

f, axs = plt.subplots(1, 2, figsize=(7, 3.5), gridspec_kw={'width_ratios': [3, 2]})
ccgps, null_ccgps = dec.CCGP(plot=True, ax=axs[1])
perfs, null_perfs = dec.decode(0.8, plot=True, non_semantic=True, ax=axs[0])
plt.suptitle('Visual area')
plt.savefig('./img/visual_decoding.png')
```
